```python
def qram(problem_set):
    qc = QuantumCircuit(20)
    
    for index, item in enumerate(problem_set):
        # Index Binary
        index_binary = "{0:04b}".format(index)
        # print(index_binary, item)
        for i_index, qubit in enumerate(index_binary):
            if qubit == '0':
                qc.x(i_index)
        for ii_index, ii_item in enumerate(item):
            # print(ii_index, ii_item)
            row = int(ii_item[0])
            column = int(ii_item[1])
            # print(row, column)
            if row == 0:
                qc.mcx([0, 1, 2, 3], column + 4)
            if row == 1:
                qc.mcx([0, 1, 2, 3], column + 8)
            if row == 2:
                qc.mcx([0, 1, 2, 3], column + 12)
            if row == 3:
                qc.mcx([0, 1, 2, 3], column + 16)
        for i_index, qubit in enumerate(index_binary):
            if qubit == '0':
                qc.x(i_index)
    
    qram_gate = qc.to_gate()
    qram_gate.name = 'qRAM'
    
    return qram_gate
```

```python
def oracle_P1():
    qc = QuantumCircuit(20)
    
    oracle_P1_gate = qc.to_gate()
    oracle_P1_gate.name = 'Oracle P1'
    
    return oracle_P1_gate
```

```python
def oracle_P2():
    qc = QuantumCircuit(20)
    
    oracle_P2_gate = qc.to_gate()
    oracle_P2_gate.name = 'Oracle P2'
    
    return oracle_P2_gate
```

```python
def diffuser():
    qc = QuantumCircuit(4)
    
    qc.x(range(4))
    qc.h(3)
    qc.mcx([0, 1, 2], 3)
    qc.h(3)
    qc.x(range(4))
    
    diffuser_gate = qc.to_gate()
    diffuser_gate.name = 'Diffuser'
    
    return diffuser_gate
```

```python
def main(lights_out):
    qc = QuantumCircuit(27, 4)
    
    # Prepare Addresses. 
    # Use Hadamard for superposition of 2 address qubits into 4 addresses.
    qc.h(range(4))

#     qc.x(ancilla)
#     qc.h(ancilla)

    r = 1
    
    while(r>0):
        # qRAM
        qram_gate = qram(problem_set)
        qc.append(qram_gate, range(20))

        # Oracle_P1
        oracle_P1_gate = oracle_P1()
        qc.append(oracle_P1_gate, range(4, 24))

        # 3 Lazer Condition

        # Oracle_P2
        oracle_P2_gate = oracle_P2()
        qc.append(oracle_P2_gate, range(4, 24))

        # qRAM
        qram_gate = qram(problem_set)
        qc.append(qram_gate, range(20))

        # Diffuser
        diffuser_gate = diffuser()
        qc.append(diffuser_gate, range(4))

        r = r-1
            
    # Measure
    qc.barrier()
    qc.measure([3, 2, 1, 0], range(4))
  
    return qc
```
